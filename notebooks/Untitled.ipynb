```python
weather = pd.read_csv('../data/weather.csv')
```

```python
weather.describe()
```

```python
weather.info()
```

```python
weather.sample(5)
```

```python
weather_melt = pd.melt(weather, id_vars=['id', 'year', 'month', 'element'],
                       var_name='day',
                       value_name='temp')
```

```python
weather_melt.head(6)
```

The pivot is performed on the columns argument. For each unique value in that column, a new column will be created.

```python
weather_tidy = pd.pivot_table(weather_melt, index=['id', 'year', 'month', 'day'],
                              columns='element',
                              values='temp')
```

```python
weather_tidy
```

Flatten the heirarchal data frame by resetting the index.

```python
weather_tidy.reset_index()
```
